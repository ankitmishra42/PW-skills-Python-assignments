**Q3 Customer Purchase Analysis**
- Identify the top 5 customers based on total sales.
- Hint: Use GroupBy with aggregation.

```python
df.groupby('Customer')['Sales'].sum().nlargest(5).reset_index()
```

**Q4 Regional Sales Performance**
- Calculate the total sales generated by each region.
- Then identify:
  - Which region generated the highest revenue?

```python
# Calculate the total sales generated by each region.
salesByRegion = df.groupby('Region')['Sales'].sum().reset_index()
print(salesByRegion)

# Which region generated the highest revenue?
topRevenue = salesByRegion[salesByRegion["Sales"] == salesByRegion["Sales"].max()]
topRevenue
```

**Q5 Product Performance Analysis**
- Calculate the total quantity sold for each product.
- Which product appears to be the most popular?

```python
# Calculate the total quantity sold for each product.
quantityByProduct = df.groupby('Product')['Quantity'].sum().reset_index()
print(quantityByProduct)

# Which product appears to be the most popular?
mostPopular = quantityByProduct[quantityByProduct["Quantity"] == quantityByProduct["Quantity"].max()]
mostPopular
```

**Q6 Multi-Level Grouping**
- Group the dataset by: Region and Category
- Calculate the total sales for each combination.
- What insights can you derive from this analysis?

```python
# Group the dataset by: Region and Category
# Calculate the total sales for each combination.
df.groupby(['Region', 'Category'])['Sales'].sum().unstack()
```

What insights can you derive from this analysis?

**Q7 Pivot Table Creation**<br>
Create a pivot table showing:
- Rows → Region
- Columns → Product
- Values → Total Sales
- Which region sells the most laptops?

```python
pivot_table = df.pivot_table(index='Region', 
                             columns='Product', 
                             values='Sales', 
                             aggfunc='sum', 
                             fill_value=0)

print(pivot_table)
print(f"\nThe region that sells the most laptops is: {pivot_table['Laptop'].idxmax()}")
```

**Q8 Profitability Analysis**
- Calculate the average profit for each product category.
- Which category is most profitable?

```python
# Calculate the average profit for each product category.
avgProfitByCategory = df.groupby('Category')['Profit'].mean().reset_index()
print(avgProfitByCategory)

# Which category is most profitable?
topAvgProfitByCategory = df.groupby('Category')['Profit'].mean().idxmax()
print(f"\n{topAvgProfitByCategory} category is most profitable")
```

**Q9 Discount Impact Study, Analyze the relationship between discount and profit.**<br>
Tasks:
- Calculate the average profit for each discount level
- Identify whether higher discounts lead to lower profit

```python
# Calculate the average profit for each discount level
avgProfitByDiscountLevel = df.groupby('Discount')['Profit'].mean().reset_index()
print(avgProfitByDiscountLevel,'\n')

# Identify whether higher discounts lead to lower profit
correlation = df['Discount'].corr(df['Profit'])
if correlation < 0 :
    print('Yes higher discounts leads to lower profit')
else :
    print('No higher discounts does not lead to lower profit')
```

**Q10 End-to-End Data Wrangling Task**<br>
Perform the following workflow:
- Filter the dataset where Sales > 3000
- Group data by Region and Product
- Calculate total sales
- Reshape the result into a pivot table
- Explain the business insights obtained from this analysis.
