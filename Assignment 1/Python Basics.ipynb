**Q2: Explain the difference between List and Tuple in Python.**
- **List:** A list is mutable, which means its elements can be changed, added, or removed after creation. It is written using square brackets [] and is used when data may need to be modified.

- **Tuple:** A tuple is immutable, meaning once it is created, its elements cannot be changed. It is written using parentheses () and is used when the data should remain constant.

**Q5: Write a Python program to check whether a number is even or odd using conditional statements.**

```python
num = int(input('Enter a number: '))
if num%2 == 0:
    print(f'The number {num} an even number!')
else:
    print(f'The number {num} an odd number!')
```

**Q10: Write a Python program to count how many times a number appears in a list. Example List [2,3,4,2,5,2]**

```python
numberList = [2,3,4,2,5,2]
for number in numberList:
    count = 0
    for numtemp in numberList:
        if number == numtemp:
            count+=1
    print(f'Number {number} appears {count} times')
```
