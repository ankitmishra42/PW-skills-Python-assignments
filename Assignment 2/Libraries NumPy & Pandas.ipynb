### Libraries: NumPy & Pandas

```python
import pandas as pd
import numpy as np
# import calendar as c
```

**Q1. Load the given retail sales dataset into a Pandas DataFrame and display the first 5 rows along with the dataset information to understand its structure.**

```python
df = pd.read_csv('retail_dataset .csv')

print(df.info())
df.head(5)
```

**Q2. Inspect the dataset and identify missing values in each column. Display the total number of missing values present in the dataset.**

```python
df.isnull().sum()
```

**Q3. Replace missing values in the Quantity and Price columns with their respective mean values using Pandas.**

```python
df['Quantity'].fillna(df['Quantity'].mean(), inplace=True)

df['Price'].fillna(df['Price'].mean(), inplace=True)
```

**Q4. Remove all rows where the Product Category or Region column contains missing values.**

```python
df = df.dropna(subset=['Product Category', 'Region'])
```

**Q5. Create a new column called Revenue using NumPy, where:**
- **Revenue = Quantity × Price**
- **Display the updated dataset with the new column.**

```python
df['Revenue'] = np.multiply(df['Quantity'], df['Price'])

df.head()
```

**Q6. Using NumPy, calculate the total revenue generated from all transactions in the dataset.**

```python
print("Total Revenue:", np.sum(df['Revenue']))
```

**Q7. Group the dataset by Product Category and calculate the total revenue generated by each category.**

```python
df.groupby('Product Category', as_index=False)['Revenue'].sum()
```

**Q8. Based on the revenue calculated for each product category, identify the top 3 categories with the highest revenue and the bottom 3 categories with the lowest revenue.**

```python
revByCat = df.groupby('Product Category', as_index=False)['Revenue'].sum()

# Top three
revByCat.sort_values(by='Revenue', ascending=False).head(3)
```

```python
# Bottom three
revByCat.sort_values(by='Revenue', ascending=False).tail(3)
```

**Q9. Group the dataset by Region and calculate the total revenue generated in each region. Identify the region with the highest revenue and the region with the lowest revenue.**

```python
revByReg = df.groupby('Region', as_index=False)['Revenue'].sum()

# Region with the highest revenue
revByReg.sort_values(by='Revenue', ascending=False).head(1)
```

```python
# Region with the lowest revenue
revByReg.sort_values(by='Revenue', ascending=False).tail(1)
```

**Q10. Extract the month from the Date column using Pandas and group the dataset by month to calculate the total monthly revenue. Also, calculate the mean, median, and standard deviation of the Revenue column using NumPy**

```python
# df.isnull().sum()


# Converting Date to datetime and extracting Month
df['Date'] = pd.to_datetime(df['Date'], dayfirst=True)
df['Month'] = df['Date'].dt.month
# df['Month'] = df['Month'].apply(lambda x: c.month_name[x])

# Grouping by Month to calculating total monthly revenue
revByMonth = df.groupby('Month')['Revenue'].sum().reset_index()
print("Total Monthly Revenue:")
print(revByMonth.head())


revArray = df['Revenue'].values
revMean = np.mean(revArray)
revMedian = np.median(revArray)
revStd = np.std(revArray)


print(f"\nRevenue Statistics:")
print(f"Mean: {revMean:.2f}")
print(f"Median: {revMedian:.2f}")
print(f"Standard Deviation: {revStd:.2f}")
```
