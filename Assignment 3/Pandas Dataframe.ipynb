**Q2. Data Inspection** <br>
Perform the following inspection tasks:
- Display the first 5 rows of the dataset.
- Display the last 5 rows.
- Show the structure of the dataset using info().
- Display the shape of the dataset.

```python
df = pd.read_csv('Retail_dataset.csv')
print('\n\n The first 5 rows of the dataset: ')
print(df.head(5)) 
print('\n\n The last 5 rows: ')
print(df.tail(5))
print('\n\n Structure of the dataset: ')
print(df.info())
print(f'\n Shape of the dataset: {df.shape}\n')
```

**Q3. Column Selection** <br> 
Select and display only the following columns:
- Product
- Region
- Sales

```python
df[['Product', 'Region', 'Sales']]
```

**Q4. Indexing using .loc** <br> 
Using .loc, perform the following:
- Display the row with index 10.
- Display the Sales value of the row with index 20

```python
print(f'The row with index 10: {df.loc[10]}')
print(f'\nThe Sales value of the row with index20: {df.loc[20, 'Sales']}')
```

**Q5. Indexing using .iloc** <br> 
Using .iloc, display:
- The first 6 rows and first 4 columns.
- The Sales value from the 3rd row.

```python
print(f"First 6 rows and 4 columns: \n {df.iloc[0:6, 0:4]}\n")

print(f"\nSales value from the 3rd row:\n {df['Sales'].iloc[2:]}")
```

**Q6. Row Slicing** <br>
Display:
- The first 10 rows of the dataset.
- Rows 15 to 25.

```python
print(f'The first 10 rows of the dataset: \n {df.iloc[0:10]}\n')

print(f'\nRows 15 to 25:\n {df.iloc[15:26]}\n')
```

**Q7. Filtering Data** <br>
Filter the dataset to display:
- Rows where Region = "East".
- Rows where Sales > 1500.

```python
df[(df['Region']=='East') & (df['Sales'] > 1500)]
```

**Q8. Multiple Condition Filtering** <br>
Display rows where:
- Region is West
- Sales is greater than 1200

```python
df[(df['Region']=='West') & (df['Sales'] > 1200)]
```

**Q9. Descriptive Statistics** <br>
Generate summary statistics using: <br>
describe() <br>
From the result identify: 
- Mean sales value
- Maximum quantity sold
- Minimum sales value

```python
print(f'Mean sales value: {df['Sales'].mean()}')
print(f'Maximum quantity sold: {max(df['Quantity'])}')
print(f'Minimum sales value: {min(df['Sales'])}')

df[['Sales','Quantity']].describe()
```

**Q10. Aggregation Analysis** <br>
Using the Sales column, calculate:
- Total sales
- Average sales
- Maximum sales
- Minimum sales

```python
print(f'Total sales: {df['Sales'].mean()}')
print(f'Average sales: {df['Sales'].mean()}')
print(f'Maximum sales: {max(df['Sales'])}')
print(f'Minimum sales: {min(df['Sales'])}')
```
