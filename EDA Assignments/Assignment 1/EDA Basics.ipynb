```python
import pandas as pd
df = pd.read_csv("House Price India.csv")

print(df.info())
df.describe()
```

#### Assignment Questions 

**Q1. What is the average, median, and standard deviation of house prices?**

```python
print(f'Average House Price: {df['Price'].mean():.2f}')
print(f'Median House Price: {df['Price'].median():.2f}')
print(f'Standard deviation of House Price: {df['Price'].std():.2f}')
```

**Q2. Which number of bedrooms is most common?**

```python
print(f'Most common number of bedrooms are: {df['number of bedrooms'].mode()[0]}')
```

**Q3. Check if the price data is skewed.**

```python
#**Q3. Check if the price data is skewed.**

print(f"Skewness of Price: {df['Price'].skew():.2f}")
```

**Q4. What is the average price for each number of bedrooms?**

```python
avg_price_beds = df.groupby('number of bedrooms')['Price'].mean().sort_index()
avg_price_beds
```

**Q5.  What is the relationship between living area and price? (Use : Area without Basement).**

```python
correlation = df['Area of the house(excluding basement)'].corr(df['Price'])
print(f"Correlation: {correlation:.2f}")
```

**Q6. Identify any anomalies where houses have high prices but low areas.**

```python
# Defining anomalies as top 10% price but bottom 25% area
anomalies = df[(df['Price'] > df['Price'].quantile(0.9)) & 
               (df['Area of the house(excluding basement)'] < df['Area of the house(excluding basement)'].median())]
print(f"Number of anomalies identified: {len(anomalies)}")
```

**Q7. Compare average price based on number of floors and houses with or without waterfront.**

```python
comparison = df.groupby(['number of floors', 'waterfront present'])['Price'].mean().unstack()
comparison
```

**Q8. Identify the minimum and maximum house price. What does this indicate?**

```python
print(f"Min Price: {df['Price'].min()} | Max Price: {df['Price'].max()}")
```

**Q9. Which location (zipcode/area) has the highest average price?**

```python
top_zip = df.groupby('Postal Code')['Price'].mean().idxmax()
print(f"Highest average price Postal Code: {top_zip}")
```
