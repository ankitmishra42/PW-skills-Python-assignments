### Data Transformation

```python
import pandas as pd

# Loading the dataset
df = pd.read_csv('employee_productivity_dataset.csv')
```

**Q1. Handle Missing Values (Basic Cleaning)**<br>
**Fill missing values using the following:**<br>
- Age → Median
- Salary → Mean
- Hours_Worked_Per_Week → Median
- Performance_Score → Mean
- Display the dataset after handling missing values.

```python
# Filling missing values
df['Age'] = df['Age'].fillna(df['Age'].median())
df['Salary'] = df['Salary'].fillna(df['Salary'].mean())
df['Hours_Worked_Per_Week'] = df['Hours_Worked_Per_Week'].fillna(df['Hours_Worked_Per_Week'].median())
df['Performance_Score'] = df['Performance_Score'].fillna(df['Performance_Score'].mean())

# The dataset after handling missing values
print("Missing values after cleaning:")
print(df.isnull().sum())
df.head()
```

**Q2. Label Encoding**<br>
**Convert these columns into numeric values using Label Encoding:**<br>
- Gender
- Department
- Show the updated columns.

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

df['Gender'] = le.fit_transform(df['Gender'].astype(str))
df['Department'] = le.fit_transform(df['Department'].astype(str))

print("Updated Gender and Department columns:")
df[['Gender', 'Department']].head()
```

**Q3. One-Hot Encoding**<br>
**Apply One-Hot Encoding on:**<br>
- Work_Mode
- Location
- Display the dataset and check how many new columns are created.

```python
# Apply One-Hot Encoding using pandas get_dummies for visualization
df_encoded = pd.get_dummies(df, columns=['Work_Mode', 'Location'])

print(f"There are total {df_encoded.shape[1]} columns after One-Hot Encoding")
df_encoded.head()
```

**Q4. Normalization (Min-Max Scaling)**<br>
**Normalize the following columns:**<br>
- Salary
- Hours_Worked_Per_Week
- Use MinMaxScaler and display results.

```python
from sklearn.preprocessing import MinMaxScaler

mms = MinMaxScaler()
df_encoded[['Salary', 'Hours_Worked_Per_Week']] = mms.fit_transform(df_encoded[['Salary', 'Hours_Worked_Per_Week']])

print("Normalized Salary and Hours Worked:")
df_encoded[['Salary', 'Hours_Worked_Per_Week']].head()
```

**Q5. Standardization (Scaling)**<br>
**Apply StandardScaler on:**<br>
- Age
- Projects_Completed
- Display the transformed values.

```python
from sklearn.preprocessing import StandardScaler

ss = StandardScaler()
# Ensure Projects_Completed has no NaNs before scaling
df_encoded['Projects_Completed'] = df_encoded['Projects_Completed'].fillna(df_encoded['Projects_Completed'].median())

df_encoded[['Age', 'Projects_Completed']] = ss.fit_transform(df_encoded[['Age', 'Projects_Completed']])

print("Standardized Age and Projects Completed:")
df_encoded[['Age', 'Projects_Completed']].head()
```

**Q6. Compare Scaling Methods**<br>
**Apply both:**<br>
- MinMaxScaler
- StandardScaler
- On the Salary column.
- Show both results side by side.

```python
# Re-calculating for comparison
original_salary = df['Salary'].values.reshape(-1, 1)

compare_df = pd.DataFrame({
    'Original': original_salary.flatten(),
    'MinMax_Scaled': MinMaxScaler().fit_transform(original_salary).flatten(),
    'Standard_Scaled': StandardScaler().fit_transform(original_salary).flatten()
})

compare_df.head()
```

**Q7. Build Preprocessing Pipeline**<br>
**Create a pipeline that:**<br>
- Applies encoding to categorical columns
- Applies scaling to numerical columns
**Use:**
- ColumnTransformer
- Pipeline

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder

# Define transformers for numerical and categorical data
num_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])

cat_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='constant', fill_value='missing')),
    ('onehot', OneHotEncoder(handle_unknown='ignore', sparse_output=False))
])

# Combine into a ColumnTransformer and Pipeline
preprocessor = ColumnTransformer(transformers=[
    ('num', num_transformer, ['Age', 'Salary', 'Hours_Worked_Per_Week', 'Projects_Completed', 'Performance_Score']),
    ('cat', cat_transformer, ['Work_Mode', 'Location', 'Gender', 'Department'])
])

full_pipeline = Pipeline(steps=[('preprocessor', preprocessor)])
```

**Q8. Apply Pipeline**<br>
**Apply the pipeline on the dataset.**<br>
**Display:**
- Transformed dataset
- Shape of final dataset

```python
# Apply the pipeline
X_processed = full_pipeline.fit_transform(df)

# Display results
print(f"Final dataset shape: {X_processed.shape}")
print("Transformed data (First 2 rows):")
print(X_processed[:2])
```

**Q9. Conceptual Question**<br>
**Why is scaling important in Python?**
- Scaling ensures all features contribute equally by bringing them to a similar range. It prevents bias toward large-value features, speeds up gradient descent, and is crucial for distance-based algorithms like KNN, SVM, and PCA.

**Q10. Conceptual Question**<br>
**Why do we convert categorical data into numerical form?**
- Machine learning models require numerical input because they perform mathematical operations. Encoding categorical data (e.g., One-Hot, Label Encoding) allows models to process and learn patterns from it.
